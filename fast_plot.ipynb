```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,3,4,5,6,7])
y0 = np.array([9999,9999,9999,9999,9999,9999,9999,9999,])
y1 = np.array([0,1882,6112,9504,9504,6112,1882,0,])
y2 = np.array([0,904,4591,9203,9203,4591,904,0,])
y3 = np.array([-4,-369,106,7808,7808,106,-369,-4,])
```

```python
plt.plot(x,y0, label="w0 rect")
plt.plot(x,y1, label="w1 hann")
plt.plot(x,y2, label="w2 blackman")
plt.plot(x,y3, label="w3 flat top")
plt.grid()
plt.legend()
plt.show()
```
